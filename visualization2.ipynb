    find_shape_files()
    Ищет файлы с данными об актёрах (начинающиеся с input_debug_ и заканчивающиеся на _shapes.csv) сначала в папке shaped_shorts, если есть, иначе в текущей директории, и возвращает список найденных файлов.

    safe_parse_shape(shape_str)
    Безопасно разбирает строку с координатами формы актёра (bounding box), извлекая 4 числа (y1, x1, y2, x2), корректно обрабатывая возможные ошибки и разные форматы.

    load_and_process_shapes(file_path)
    Загружает CSV-файл с данными актёров, проверяет необходимые колонки, парсит координаты с помощью safe_parse_shape, добавляет новые столбцы с координатами рамки, шириной, высотой и площадью.

    calculate_actor_presence(shape_files, fps=24)
    Считает для каждого актёра общее количество кадров с его появлением, количество шотов, среднюю площадь bounding box и общее время присутствия в секундах, агрегируя данные из всех файлов.

    visualize_actor_presence_by_shots(stats_df)
    Строит горизонтальную столбчатую диаграмму среднего времени актёров в кадре за один шот с подписями численных значений.

    load_scene_or_shot_data(prefix)
    Загружает CSV-файл анализа сцен или шотов по префиксу в имени (scene_analysis* или shot_analysis*), проверяет обязательные колонки, сортирует данные по времени.

    visualize_duration_distribution(df, title)
    Строит гистограмму распределения длительностей сцен или шотов (по колонке duration_seconds).

    visualize_actor_timeline(shape_files, scene_df)
    Визуализирует временную шкалу появления актёров в кадре (по кадрам и времени), разбивая по шотам, а также отображает вертикальные линии с подписями границ сцен.

    main_analysis()
    Основная функция: ищет и загружает файлы с данными актёров, сцен и шотов; рассчитывает статистику присутствия актёров; выводит информацию; строит гистограммы и визуализирует временные линии актёров с обозначением сцен.


```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
import re
from collections import defaultdict
from tqdm import tqdm

def find_shape_files():
    """Поиск файлов с данными актёров в shaped_shorts или текущей директории"""
    search_dirs = []
    if os.path.exists('shaped_shorts'):
        search_dirs.append('shaped_shorts')
    search_dirs.append('.')  # Текущая директория

    shape_files = []
    for dir_path in search_dirs:
        found_files = [
            os.path.join(dir_path, f)
            for f in os.listdir(dir_path)
            if f.startswith('input_debug_') and f.endswith('_shapes.csv')
        ]
        if found_files:
            shape_files.extend(found_files)
            if dir_path == 'shaped_shorts':
                print("Файлы с актерами найдены в папке shaped_shorts")
            else:
                print("Файлы с актерами найдены в текущей директории")
            break

    if not shape_files:
        print("Файлы с данными об актерах не найдены ни в папке shaped_shorts, ни в текущей директории")

    return shape_files

def safe_parse_shape(shape_str):
    """Надёжный парсинг координат из строки shape"""
    try:
        if pd.isna(shape_str) or not isinstance(shape_str, str):
            return [0, 0, 0, 0]

        raw_coords = re.sub(r'[()]', '', shape_str).split(',')
        coords = [c.strip() for c in raw_coords if c.strip() != '']

        if len(coords) != 4:
            return [0, 0, 0, 0]

        parsed_coords = []
        for c in coords:
            # Пробуем преобразовать в float и затем в int
            val = int(float(c))
            parsed_coords.append(val)
        return parsed_coords
    except Exception as e:
        # Можно раскомментировать для отладки:
        # print(f"Ошибка при парсинге shape: {shape_str}, ошибка: {e}")
        return [0, 0, 0, 0]

def load_and_process_shapes(file_path):
    """Безопасная загрузка и обработка файла с данными актёров"""
    try:
        df = pd.read_csv(file_path)
        required_cols = ['frame', 'shape', 'name']
        missing = [col for col in required_cols if col not in df.columns]
        if missing:
            print(f"Файл {os.path.basename(file_path)} отсутствуют колонки: {missing}")
            return None

        shape_data = df['shape'].apply(safe_parse_shape).apply(pd.Series)
        if shape_data.shape[1] != 4:
            print(f"Файл {os.path.basename(file_path)} содержит некорректные данные в колонке 'shape'")
            return None

        df[['y1', 'x1', 'y2', 'x2']] = shape_data
        df['width'] = df['x2'] - df['x1']
        df['height'] = df['y2'] - df['y1']
        df['area'] = df['width'] * df['height']

        return df.dropna(subset=['frame', 'name'])

    except Exception as e:
        print(f"Ошибка при обработке {os.path.basename(file_path)}: {e}")
        return None

def calculate_actor_presence(shape_files, fps=24):
    """Расчет статистики присутствия актеров"""
    actor_stats = defaultdict(lambda: {
        'total_frames': 0,
        'total_shots': 0,
        'avg_area': [],
        'shot_list': []
    })

    for file in tqdm(shape_files, desc="Анализ файлов с актерами"):
        df = load_and_process_shapes(file)
        if df is None:
            continue

        # Номер шота из имени файла
        basename = os.path.basename(file)
        shot_num_search = re.search(r'input_debug_(\d+)_shapes\.csv', basename)
        shot_num = shot_num_search.group(1) if shot_num_search else basename

        for actor, group in df.groupby('name'):
            actor_stats[actor]['total_frames'] += len(group)
            actor_stats[actor]['avg_area'].append(group['area'].mean())
            if shot_num not in actor_stats[actor]['shot_list']:
                actor_stats[actor]['total_shots'] += 1
                actor_stats[actor]['shot_list'].append(shot_num)

    stats_df = pd.DataFrame.from_dict(actor_stats, orient='index')
    stats_df['total_time'] = stats_df['total_frames'] / fps
    stats_df['avg_area'] = stats_df['avg_area'].apply(lambda x: np.mean(x) if x else 0)
    stats_df = stats_df.sort_values('total_time', ascending=False)

    return stats_df[['total_time', 'total_shots', 'avg_area', 'shot_list', 'total_frames']]

def visualize_actor_presence_by_shots(stats_df):
    """Гистограмма среднего времени актёров по шотам"""
    if stats_df.empty:
        print("Нет данных для визуализации присутствия по шотам")
        return

    stats_df['time_per_shot'] = stats_df['total_time'] / stats_df['total_shots']
    plt.figure(figsize=(12, 6))
    actors = [str(a) for a in stats_df.index]
    times = stats_df['time_per_shot']
    bars = plt.barh(actors, times, color=plt.cm.tab20.colors[:len(actors)])
    plt.title('Среднее время присутствия актеров на один шот (секунды)')
    plt.xlabel('Время (секунды)')
    plt.ylabel('Актеры')
    plt.grid(axis='x', alpha=0.3)
    for bar, val in zip(bars, times):
        width = bar.get_width()
        plt.text(width + max(times)*0.01, bar.get_y() + bar.get_height()/2,
                 f'{val:.1f}', va='center')
    plt.tight_layout()
    plt.show()

def load_scene_or_shot_data(prefix):
    """Загрузка CSV с анализом сцен или шотов"""
    file_name = next(
        (f for f in os.listdir('.') if f.lower().startswith(prefix) and f.endswith('.csv')),
        None
    )
    if not file_name:
        print(f"Не найден файл анализа {prefix}")
        return None
    try:
        df = pd.read_csv(file_name)
        required_cols = ['start_time', 'end_time', 'duration_seconds']
        if not all(col in df.columns for col in required_cols):
            print(f"Файл {file_name} не содержит необходимых колонок {required_cols}")
            return None
        df['mid_time'] = (df['start_time'] + df['end_time']) / 2
        return df.sort_values('start_time').reset_index(drop=True)
    except Exception as e:
        print(f"Ошибка при загрузке файла {file_name}: {e}")
        return None

def visualize_duration_distribution(df, title):
    """Гистограмма распределения длительности сцен или шотов"""
    if df is None or df.empty:
        print(f"Нет данных для визуализации распределения длительности {title}")
        return
    plt.figure(figsize=(10, 6))
    plt.hist(df['duration_seconds'], bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Распределение длительности: {title}')
    plt.xlabel('Длительность (секунды)')
    plt.ylabel('Количество')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def visualize_actor_timeline(shape_files, scene_df):
    """Визуализация временной линии появления актеров (шоты) и границ сцен"""
    if not shape_files:
        print("Нет данных для построения временной линии актеров (нет файлов шотов)")
        return

    all_shots = []
    for file in shape_files:
        try:
            basename = os.path.basename(file)
            shot_num_search = re.search(r'input_debug_(\d+)_shapes\.csv', basename)
            shot_num = int(shot_num_search.group(1)) if shot_num_search else -1
            df = load_and_process_shapes(file)
            if df is not None:
                df['shot_num'] = shot_num
                all_shots.append(df)
        except Exception as e:
            print(f"Ошибка при обработке файла {file}: {e}")
            continue

    if not all_shots:
        print("Нет корректных данных по шотам для построения временной линии")
        return

    combined_df = pd.concat(all_shots)
    if combined_df.empty:
        print("Данные для визуализации пусты")
        return

    actors = combined_df['name'].unique()
    if len(actors) == 0:
        print("Нет данных об актерах для визуализации")
        return

    colors = plt.cm.tab20(np.linspace(0,1,len(actors)))
    plt.figure(figsize=(15,8))
    for i, actor in enumerate(actors):
        actor_data = combined_df[combined_df['name'] == actor]
        for _, row in actor_data.iterrows():
            plt.hlines(y=actor, xmin=row['frame']/24, xmax=(row['frame']+1)/24,
                       colors=colors[i], linewidth=5)

    plt.title('Временная линия появления актеров по шотам')
    plt.xlabel('Время (секунды)')
    plt.ylabel('Актеры')
    plt.yticks(range(len(actors)), actors)

    if scene_df is not None:
        for scene_idx, scene in scene_df.iterrows():
            plt.axvline(x=scene['start_time'], color='red', linestyle='--', alpha=0.5)
            plt.axvline(x=scene['end_time'], color='red', linestyle='--', alpha=0.5)
            mid = (scene['start_time'] + scene['end_time'])/2
            plt.text(mid, len(actors)+0.5, f"Сцена {scene_idx}", ha='center', color='red')

    plt.tight_layout()
    plt.show()

def main_analysis():
    shape_files = find_shape_files()
    if not shape_files:
        return

    # Загружаем данные сцен и шотов
    scene_df = load_scene_or_shot_data('scene_analysis')
    shot_df = load_scene_or_shot_data('shot_analysis')

    # Анализ присутствия актёров
    stats_df = calculate_actor_presence(shape_files)
    if stats_df.empty:
        print("Ошибка: нет данных для анализа актёров")
        return

    print("\n=== Статистика присутствия актеров ===")
    print(stats_df[['total_time', 'total_shots', 'avg_area']])

    # Визуализация
    if scene_df is not None:
        visualize_duration_distribution(scene_df, "Сцены")
    if shot_df is not None:
        visualize_duration_distribution(shot_df, "Шоты")

    visualize_actor_timeline(shape_files, scene_df)
    visualize_actor_presence_by_shots(stats_df)

if __name__ == "__main__":
    main_analysis()
```
